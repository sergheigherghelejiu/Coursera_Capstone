```python
#print(myTable)
html = urlopen(url)
sp = BeautifulSoup(html, 'lxml')
mytb2 =  sp.findAll(  "td")
print(mytb2)
```

```python
cleantext = BeautifulSoup(str(mytb2), "lxml").get_text()
print(cleantext)
```

```python
import  pandas as pd
import io

my_list = cleantext.split("\n")
#print(my_list)
myTable = []
for row in my_list:
    myTable.append(row[1:].split(","))
df2 = pd.DataFrame(myTable)
print(df2)
```

```python
df2.columns = ["PostalCode", "Borough", "Neighborhood"]
print(df2)
```

```python
dfc = df2.dropna()
df4 =  dfc[dfc["Borough"].str.strip() != 'Not assigned']
df4
```

```python
df4.Neighborhood = df4.loc[df4.Neighborhood.str.strip() == 'Not assigned', 'Neighborhood'] = df4["Borough"]
#df4['Neighborhood'] = np.where(df4.Neighborhood.str.strip() == 'Not assigned',df4.Borough ,df4.Neighborhood)
```
