```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python

soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
myTable = soup.findAll("table", {"class": "wikitable sortable"})
```

```python
#print(myTable)
sp = BeautifulSoup(html, 'lxml')
mytb2 =  sp.findAll(  "td")
print(mytb2)
```

```python
cleantext = BeautifulSoup(str(mytb2), "lxml").get_text()
print(cleantext)
```

```python
import  pandas as pd
import io
df = pd.DataFrame(cleantext)
print(df)
```

```python
my_list = cleantext.split("\n")
#print(my_list)
myTable = []
for row in my_list:
    myTable.append(row[1:].split(","))
df2 = pd.DataFrame(myTable)
print(df2)
```
